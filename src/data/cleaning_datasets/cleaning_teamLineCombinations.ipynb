##### As only one single value exists, we can remove the `situation` column.

#### However, also note that we need to do this for every single lines dataset and then combine all those cleaned datasets into one dataset to fit the database schema. Thus, we will make a single function to accomplish all these tasks.

```python
# making function

def cleaning_dataset(dataframe, lineType): 
    df = pd.read_csv(dataframe)

    # filtering to only get Toronto Maple Leafs line combinations
    if lineType == 'regular':
        filters = (df['team'].values == 'TOR') & (df['games_played'] >= 10)
    else:
        filters = df['team'].values == 'TOR'

    df = df[filters]

    # dropping useless columns
    dropped_columns = [ 
        'season',
        'lineId',
        'situation',
        'team',
        'pairing'
    ]
    df = df.drop(columns=dropped_columns, errors='ignore')

    return df.to_dict(orient='records')
```

```python
# get a list of all seasons
year = date.today().year
years = [2016 + x for x in range(year - 2016)]
seasons = [f'{years[i]}{years[i]+1}' for i in range(len(years))]
print(seasons)

# get list of all line tyoes
line_datasets_types = list(map(lambda x: x.split("_")[1], os.listdir("../scraped_csv/team_lines/")))
print(line_datasets_types)
```

```python
# do this for every line dataset
all_lines_stats = {}
for season in seasons:
    current_line_stats = {}
    for lineType in line_datasets_types:
        # print(f'{player}_{season}')
        if os.path.isfile(f"../scraped_csv/team_lines/{season}_{lineType}_lines.csv"):
            try:
                current_line_stats[f'{lineType}'] = cleaning_dataset(f"../scraped_csv/team_lines/{season}_{lineType}_lines.csv", lineType)
            except EmptyDataError:
                current_line_stats[f'{season}'] = {'stats': 'no stats for this year'}
    all_lines_stats[f'{season}'] = current_line_stats
all_lines_stats
df = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in all_lines_stats.items() ]))
df
```

### Lets finally export the cleansed dataframe into a csv file

```python
title = '../cleaned_csv/teamLinesAllCombinationsCleaned.csv'
df.to_csv(title, index=False)
```
