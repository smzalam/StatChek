### Import Dependencies

```python
import pandas as pd
import numpy as np
```

### Read CSV file.

```python
df = pd.read_csv("../scraped_csv/team_details/teamAllDetails.csv")
df
```

### Hard to see all columns, so lets see the transposed dataframe instead.

```python
df = df.transpose()
df
```

### Looking at database schema, we only need some of these columns, so I will now remove columns that are not needed.

```python
dropped_columns = [
    'link', 
    'teamName', 
    'shortName', 
    'franchiseId', 
    'active', 
    'venue.link', 
    'venue.city', 
    'venue.timeZone', 
    'division.name', 
    'division.nameShort', 
    'division.link', 
    'division.abbreviation',
    'conference.name', 
    'conference.link', 
    'franchise.franchiseId', 
    'franchise.teamName', 
    'franchise.link',
    'venue.id'
]
df = df.drop(dropped_columns)
df
```

### It will be easier to see full list and other blocks of code horizontally so we once again transpose the dataframe.

```python
df = df.transpose()
df
```

### Let's rename these columns to something more meaningful.

```python
renamed_columns = {
    'id': 'team_id', 
    'name': 'name',
    'abbreviation': 'abbreviation',
    'locationName': 'location',
    'firstYearOfPlay': 'inaugaration_year', 
    'officialSiteUrl': 'url', 
    'venue.name': 'venue', 
    'division.id': 'division_id', 
    'conference.id':'conference_id', 
}
df = df.rename(columns=renamed_columns)
df
```

```python
teams_df = df[['team_id', 'name']]
teams_df
```

### Let's check if the csv shows up as we want to in the format it will be inserted into the database in

```python
payload_teams_info = df.to_dict(orient='records')
payload_teams_info
```

```python
payload_teams_listings = teams_df.to_dict(orient='records')
payload_teams_listings
```

### Lets finally export the cleansed dataframe into a csv file

```python
title = '../cleaned_csv/teamAllDetailsCleaned.csv'
df.to_csv(title, index=True)
```

```python
title = '../cleaned_csv/teamAllListingsCleaned.csv'
teams_df.to_csv(title, index=True)
```
